```python
def read_files(path, dos, donts, kind):
    data = None
    for root, dirs, files in os.walk(path):
        for file in files:
            if file == ".DS_Store" or not file.startswith(kind) or not any([kind in file for kind in dos]) or any([kind in file for kind in donts]):
                continue
            try:
                d = pd.read_csv(os.path.join(root, file), sep=";")
                print(file, "success")
            except:
                print(file, "failure")
            d["type"] = file.split(".")[0]
            exp = file.split(".")[3]
            d["exp"] = exp
            d["run"] = file.split(".")[2]
            d["evolver"] = file.split(".")[1]
            configuration = file.split(".")[4]
            d["configuration"] = configuration
            d["shape"] = file.split(".")[5]
            d["sensors"] = "uniform" if file.split(".")[6] == "csv" else file.split(".")[6]
            d["task"] = "locomotion" if file.split(".")[6] == "csv" else file.split(".")[6]
            if data is None:
                data = pd.DataFrame(columns=d.columns)
            data = data.append(d)
    return data
```

```python
def plot_fitness(data, conf_to_label, by="configuration", column="best→fitness→as[Outcome]→velocity"):
    fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=2)
    data["iterations"] = data["iterations"].astype(np.int32)
    
    for (conf, shape), traj in data.groupby([by, "shape"]):
        col = 0 if "biped" in shape else 1
        median = traj.groupby(traj.iterations)[column].median()
        err = traj.groupby(traj.iterations)[column].std()
        color="red" if "tanh" in conf else "blue"
        linestyle = "solid" if "none" in conf else "dashed"
        axes[col].plot(median, label=conf_to_label[conf], color=color, linestyle=linestyle)
        axes[col].fill_between(traj.groupby(traj.iterations)["iterations"].median(), median - err, median + err, alpha=0.15, color=color)
        axes[col].set_xlabel("iterations", fontsize=15)
        axes[col].set_ylabel("velocity", fontsize=15)
        axes[col].legend(fontsize=15)
        axes[col].set_title(shape, fontsize=15)
        
    fig.tight_layout()
```

```python
def plot_boxplots(data, conf_to_label, by="temp_configuration"):
    fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=2)
    if by == "temp_configuration":
        data["temp_configuration"] = data.apply(lambda row: conf_to_label[row["configuration"]], axis=1)
        data.sort_values(by="temp_configuration")
    
    for shape, traj in data.groupby(["shape"]):
        d = {}
        colors = []
        i = 0
        col = 0 if "biped" in shape else 1
        for (conf, exp), t in traj.groupby([by, "exp"]):
            t = t[t["iterations"] == t["iterations"].max()]
            d["|".join([conf, exp if exp != "baseline" else "MLP"])] = t["best→fitness→as[Outcome]→velocity"]
            colors.append("blue" if i % 2 == 0 else "red")
            i += 1
        box = axes[col].boxplot(d.values())
        for _, line_list in box.items():
            i = 0
            for line in line_list:
                line.set_color("blue" if "MLP" in list(d.keys())[int(i / (len(line_list) / len(colors)))] else "red")
                i += 1
        axes[col].set_xticklabels(d.keys(), fontsize=15)
        axes[col].set_xlabel("experiment", fontsize=15)
        axes[col].set_ylabel("velocity", fontsize=15)
        axes[col].set_ylim(0.0, 10.0)
        axes[col].set_title(shape, fontsize=15)
        axes[col].grid()

    fig.suptitle("")
    fig.tight_layout()
```

```python
def plot_validation_boxplots(data, conf_to_label, terrain=None, by="temp_configuration"):
    fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=2)
    if terrain is not None:
        data = data[data["validation.terrain"].apply(lambda row: terrain in row)]
    if by == "temp_configuration":
        data["temp_configuration"] = data.apply(lambda row: conf_to_label[row["configuration"]], axis=1)
        data.sort_values(by="temp_configuration")
    
    for shape, traj in data.groupby(["shape"]):
        d = {}
        colors = []
        i = 0
        col = 0 if "biped" in shape else 1
        for (conf, exp), t in traj.groupby([by, "exp"]):
            if by == "temp_configuration":
                d[exp if ("baseline" not in exp and "baseline" not in conf) else "MLP"] = t["outcome.velocity"]
            else:
                d["|".join([conf, exp if ("baseline" not in exp and "baseline" not in conf) else "MLP"])] = t["outcome.velocity"]
            colors.append("blue" if i % 2 == 0 else "red")
            i += 1
        box = axes[col].boxplot(d.values())
        for _, line_list in box.items():
            i = 0
            for line in line_list:
                line.set_color("blue" if "MLP" in list(d.keys())[int(i / (len(line_list) / len(colors)))] else "red")
                i += 1
        axes[col].set_xticklabels(d.keys(), fontsize=15)
        axes[col].set_xlabel("experiment", fontsize=15)
        axes[col].set_ylabel("velocity", fontsize=15)
        axes[col].set_ylim(0.0, 10.0)
        axes[col].set_title(shape, fontsize=15)
        axes[col].grid()
        
    fig.suptitle("")
    fig.tight_layout()
```

```python
def plot_fine_tuning(data, t=50, column="best→fitness→as[Outcome]→velocity"):
    fig, axes = plt.subplots(figsize=(20, 10), nrows=2, ncols=2)
    for (kind, exp, shape, config), traj in data.groupby(["type", "exp", "shape", "configuration"]):
        col = 0 if "biped" in shape else 1
        row = 0 if ("6x4" in shape or "14x2" in shape) else 1
        linestyle = "solid" if kind == "finetune" else "dashed"
        label = "evo" if kind == "best" else "fine-tune"
        color = "red" if "tanh" in config else "blue"
        median = traj.groupby(traj.iterations)[column].median()
        err = traj.groupby(traj.iterations)[column].std()
        axes[row][col].plot(median, label="|".join([label, exp if exp != "baseline" else "MLP"]), linestyle=linestyle, color=color)
        axes[row][col].fill_between(list(range(len(median))), median - err, median + err, alpha=0.15, color=color)
        axes[row][col].set_title(shape, fontsize=15)
        
    for row in range(len(axes)):
        for col in range(len(axes[row])):
            axes[row][col].set_xlabel("iterations", fontsize=15)
            axes[row][col].set_ylabel("velocity", fontsize=15)
            axes[row][col].legend(fontsize=15)
    fig.suptitle("")
    fig.tight_layout()
```

```python
def plot_alpha_eta(data):
    fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=2)
    data["iterations"] = data["iterations"].astype(np.int32)
    
    for shape, traj in data.groupby(["shape"]):
        col = 0 if "biped" in shape else 1
        for column in ["best→solution→alpha", "best→solution→eta"]:
            median = traj.groupby(traj.iterations)[column].median()
            err = traj.groupby(traj.iterations)[column].std()
            axes[col].plot(median, label=column.split("→")[2])
            axes[col].fill_between(traj.groupby(traj.iterations)["iterations"].median(), median - err, median + err, alpha=0.15)
        axes[col].set_xlabel("iterations", fontsize=15)
        axes[col].set_ylabel("value", fontsize=15)
        axes[col].legend(fontsize=15)
        axes[col].set_title(shape, fontsize=15)
        
    fig.tight_layout()
```
